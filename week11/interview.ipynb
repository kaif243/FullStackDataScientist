# 1. What is Python and why is it popular in Data Science?
### Python is a high-level programming language that is popular in data science because of its simplicity and powerful libraries like NumPy and Pandas.

# 2. Difference between List, Tuple, Set, and Dictionary

### List is ordered and mutable, Tuple is ordered and immutable, Set is unordered with unique values, and Dictionary stores data in key-value pairs.

# 3. What is a Lambda function?

### A lambda function is an anonymous one-line function used for short operations.

# 4. What is map(), filter(), reduce()?

### map() applies a function to all elements, filter() selects elements based on a condition, and reduce() reduces elements to a single value.

# 5. Difference between NumPy array and Python list?

### NumPy arrays are faster and support vectorized operations, while Python lists are slower and more flexible.

# 6. What is Pandas used for?

### Pandas is used for data manipulation, cleaning, and analysis using DataFrames.

# 7. How do you handle missing values?

### Missing values can be handled by removing them or replacing them with mean, median, or mode.

# 8. Difference between loc and iloc

### loc accesses data using labels, while iloc accesses data using index positions.

# 9. What is Mean, Median, Mode?

### Mean is the average, Median is the middle value, and Mode is the most frequent value.

# 10. What is Variance and Standard Deviation?

### Variance measures data spread, and standard deviation is the square root of variance.

# 11. What is Correlation?

### Correlation measures the relationship between two variables.

# 12. What is Machine Learning?

### Machine learning is a technique that allows systems to learn from data without being explicitly programmed.

# 13. Types of Machine Learning

### Supervised, Unsupervised, and Reinforcement learning.

# 14. Difference between Regression and Classification

### Regression predicts continuous values, while classification predicts categories.

# 15. What is Linear Regression?

### Linear regression models the relationship between input and output using a straight line.

# 16. What are assumptions of Linear Regression?

### Linearity, independence, homoscedasticity, and normality.

# 17. What is R² score?

### R² score measures how well the model explains the variance in the data.

# 18. What is Logistic Regression?

### Logistic regression is a classification algorithm used to predict binary outcomes.

# 19. Why sigmoid function is used?

### Sigmoid function converts values into probabilities between 0 and 1.

# 20. Difference between Linear and Logistic Regression

### Linear regression predicts continuous values, while logistic regression predicts probabilities or classes.

# 21. What is Margin in SVM?

### Margin is the distance between the decision boundary and nearest data points.

# 22. What are Support Vectors?

### Support vectors are the data points closest to the decision boundary.

# 23. What is Kernel Trick?

###  Kernel trick transforms data into higher dimensions to make it linearly separable.

# 24. Types of Kernels

### Linear, Polynomial, RBF, and Sigmoid kernels.

# 25. What is C parameter in SVM?

### C controls the trade-off between margin size and classification error.

# 26. What is Train-Test Split?

### Train-test split divides data into training and testing sets.

# 27. Difference between Batch and Streaming

### Batch processes large data at once, while streaming processes data in real time.

# 28. What is ETL?

### ETL stands for Extract, Transform, and Load of data.

# 29. Difference between Bias and Variance

### Bias is error due to oversimplification, and variance is error due to overfitting.
